```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
theta=np.linspace(0,2*np.pi)
x=2*np.cos(theta)+np.cos(2*theta)
y=2*np.sin(theta)-np.sin(2*theta)
```

```python
plt.figure(dpi=200)
plt.plot(x,y)
plt.show()
```

# b)

```python
theta=np.linspace(0,10*np.pi)
r=theta**2

x=r*np.cos(theta)
y=r*np.sin(theta)
```

```python
plt.figure(dpi=200)
plt.plot(x,y)
plt.show()
```
